# EDA (Exploratory data Analysis)

# objective : To Analysis The Dataset,understand patters,and extract meaningfull insights.

```python
import pandas as pd
df = pd.read_csv("hackernews_dataset.csv")
df.head()
df.info()
df.describe()
```

# Analyze Patterns

```python
df["Title_Length"] = df["Title"].apply(len)

df.head()
```

# Find Insights

```python
print("Total Titles:", len(df))
print("Average Length:", df["Title_Length"].mean())
print("Max Length:", df["Title_Length"].max())
```

# Detect Data Issues

```python
print(df.isnull().sum())
```

# Hypothesis Testing 

# Hypothesis is used to checking Longer news titles have higher importance or frequency.

# Group Analysis (Simple Logic)

```python
avg_length = df["Title_Length"].mean()
print("Average Title Length:", avg_length)
long_titles = df[df["Title_Length"] > avg_length]
short_titles = df[df["Title_Length"] <= avg_length]

print("Long Titles:", len(long_titles))
print("Short Titles:", len(short_titles))
```

# Visualization

```python
import matplotlib.pyplot as plt

plt.hist(df["Title_Length"])
plt.title("Distribution of Title Length")
plt.xlabel("Length")
plt.ylabel("Count")
plt.show()
```

# Ask Meaningful Questions

# 1. How many news titles are there?

```python
total_titles = len(df)
print("Total number of news titles:", total_titles)
```

# 2. What is the average length of titles?

```python
df["Title_Length"] = df["Title"].apply(len)

avg_length = df["Title_Length"].mean()
print("Average title length:", avg_length)
```

# 3. Are there long or short titles?

```python
long_titles = df[df["Title_Length"] > avg_length]
short_titles = df[df["Title_Length"] <= avg_length]

print("Number of long titles:", len(long_titles))
print("Number of short titles:", len(short_titles))
```
