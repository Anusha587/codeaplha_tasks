# TASK 3: Data Visualization

# 1. Transform raw data into visual formats
Explanation:
Raw data (like scraped news titles) is hard to understand. So we convert it into charts like bar graphs, pie charts, etc.

```python
import requests
from bs4 import BeautifulSoup

url = "https://news.ycombinator.com/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
titles = soup.find_all("span", class_="titleline")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
titles_list = [title.text for title in titles]
df = pd.DataFrame({
    "Title": titles_list
})
df["Length"] = df["Title"].apply(len)
plt.figure()
plt.bar(range(len(df)), df["Length"])
plt.xlabel("Index")
plt.ylabel("Title Length")
plt.title("Title Length Distribution")
plt.show()
```

# 2. Use tools like Matplotlib / Seaborn
Explanation:
Libraries like matplotlib help us create graphs easily.

```python
plt.figure()
plt.hist(df["Length"])
plt.title("Distribution of Title Length")
plt.xlabel("Length")
plt.ylabel("Frequency")
plt.show()
```

 # 3. Design visuals that enhance understanding
Explanation:
Good visuals make data easy to understand quickly.

```python
short = sum(df["Length"] < 50)
long = sum(df["Length"] >= 50)

labels = ["Short Titles", "Long Titles"]
sizes = [short, long]

plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Short vs Long Titles")
plt.show()
```
